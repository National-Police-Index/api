```python
import json
```

```python
filepath = "../data/output/auto_matched.jsonl"

post_ids = set()
total_rows = 0

with open(filepath, 'r') as f:
    for line in f:
        row = json.loads(line.strip())
        total_rows += 1
        post_ids.add(row["post_match"]["post_person_nbr"])

print(f"Total rows: {total_rows}")
print(f"Unique POST IDs: {len(post_ids)}")
print(f"Duplicate rate: {(total_rows - len(post_ids)) / total_rows * 100:.1f}%")
```
