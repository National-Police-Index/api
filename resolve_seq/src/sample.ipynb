```python
import pandas as pd
import random
from openpyxl import load_workbook

def sample_employment_histories(
    input_file: str = "data/output/matched_clean_no_conflicts.xlsx",
    output_file: str = "data/output/matched_employment_histories_no_conflicts_sample_100.xlsx",
    sample_size: int = 100
):
    """
    Randomly sample tabs from an Excel file and create a new file with the sample.
    
    Args:
        input_file: Path to the input Excel file
        output_file: Path to save the sampled Excel file
        sample_size: Number of tabs to sample (default: 100)
    """
    # Load the workbook to get all sheet names
    wb = load_workbook(input_file, read_only=True)
    all_sheets = wb.sheetnames
    wb.close()
    
    print(f"Total sheets in file: {len(all_sheets)}")
    
    # Sample sheets (or take all if fewer than sample_size)
    num_to_sample = min(sample_size, len(all_sheets))
    sampled_sheets = random.sample(all_sheets, num_to_sample)
    
    print(f"Sampling {num_to_sample} sheets...")
    
    # Create new Excel file with sampled sheets
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for sheet_name in sampled_sheets:
            # Read the entire sheet
            df = pd.read_excel(input_file, sheet_name=sheet_name, header=None)
            # Write to new file
            df.to_excel(writer, sheet_name=sheet_name, index=False, header=False)
    
    print(f"Created sample file with {num_to_sample} sheets: {output_file}")
    return sampled_sheets



sampled = sample_employment_histories()
print(f"\nSampled sheets: {sampled[:10]}...")  # Print first 10
```
